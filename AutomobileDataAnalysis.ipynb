### Now lets analyse continuous features

```python
continuous_feature = [feature for feature in num_feature if feature not in disc_feature]
print('There are {} continuous features..'.format(len(continuous_feature)))
continuous_feature
```

### lets see the distribution of continuous features

```python
for feature in continuous_feature:
    sns.distplot(df[feature], bins=30, color='m')
    plt.xlabel(feature)
    plt.title('Distribution of '+feature)
    plt.show()
```

```python
for feature in continuous_feature:
    d = df.copy()
    d[feature] = np.log(d[feature])
    sns.distplot(d[feature], bins=30, color='m')
    plt.xlabel(feature)
    plt.title('Distribution of '+feature)
    plt.show()
```

### Working with outliers

```python
for feature in continuous_feature:
    df.boxplot(column=feature)
    plt.title(feature)
    plt.show()
```

```python
for feature in continuous_feature:
    l = np.mean(df[feature])-2*np.std(df[feature])
    h = np.mean(df[feature])+2*np.std(df[feature])
    print('sd Range of ', feature, ': ', np.round(l,4), ' to ', np.round(h,4))
    outliers = []
    values = []
    for val in df[feature].values:
        if val>h or val<l:
            outliers.append(1)
            values.append(val)
        else:
            outliers.append(0)
    d = df.copy()
    d[feature+'_sd_outliers'] = outliers
    print('There are {} sd outliers in '.format(len(values)), feature)
    print(values)
    sns.distplot(df[feature], bins=25, color='m')
    #plt.axvline(x=minimum, ymin=0, ymax=1, fillstyle='left', color='m')
    #plt.axvline(x=maximum, ymin=0, ymax=1, fillstyle='right', color='m')
    plt.axvspan(h, df[feature].max(), color='y')
    plt.axvspan(df[feature].min(), l, color='y')
    plt.title(feature+' with outliers')
    plt.xlabel(feature)
    plt.show()
    
```

```python
for feature in continuous_feature:
    q3 = df[feature].quantile(0.75)
    q1 = df[feature].quantile(0.25)
    iqr = q3-q1
    l = q1-1.5*iqr
    h = q3+1.5*iqr
    print('sd Range of ', feature, ': ', np.round(l,4), ' to ', np.round(h,4))
    outliers = []
    values = []
    for val in df[feature].values:
        if val>h or val<l:
            outliers.append(1)
            values.append(val)
        else:
            outliers.append(0)
    d = df.copy()
    d[feature+'_sd_outliers'] = outliers
    print('There are {} sd outliers in '.format(len(values)), feature)
    print(values)
    sns.distplot(df[feature], bins=25, color='m')
    #plt.axvline(x=minimum, ymin=0, ymax=1, fillstyle='left', color='m')
    #plt.axvline(x=maximum, ymin=0, ymax=1, fillstyle='right', color='m')
    plt.axvspan(h, df[feature].max(), color='y')
    plt.axvspan(df[feature].min(), l, color='y')
    plt.title(feature+' with outliers')
    plt.xlabel(feature)
    plt.show()
    
```
